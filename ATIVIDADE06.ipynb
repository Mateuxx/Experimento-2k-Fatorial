## ATIVIDADE 06 - Experimento do tipo 2^k fatorial

Os experimentos fatoriais 2^k são um tipo especial de experimento fatorial completo. Nestes experimentos, cada fator tem apenas dois níveis, e todos os níveis de todos os fatores são testados. O número total de experimentos 2^k é 2^k, onde k é o número de fatores


Os experimentos fatoriais 2^k são um método eficiente para estudar o efeito de vários fatores. Eles são relativamente simples de planejar e executar, e requerem um número relativamente pequeno de experimentos.

O objetivo é determinar quais fatores afetam significativamente o resultado do experimento e em que medida.

```python
def calcular_sst(tabela, valores):
    y_values = [linha[-1] for linha in tabela if len(linha) > 4]
    y_values = [int(value) for value in y_values if isinstance(value, (int, float))]
    y_mean = sum(y_values) / len(y_values)
    total = sum([(yi - y_mean) ** 2 for yi in y_values])
    print(f'valores {valores}')
    sst = []
    for valor in valores:
      sst.append(len(y_values)*(valor**2))
    print(sst)
    return sst

def calcular_efeitos(fatores, sst):
    efeitos = {}
    total_sst = sum(sst)

    for fator, sst_valor in zip(fatores, sst):
        efeitos[fator] = (sst_valor / total_sst) * 100

    return efeitos

def imprimir_efeitos(dicionario):
    for chave, valor in dicionario.items():
        porcentagem = round(valor, 2)  # Multiplica por 100 e arredonda para 2 casas decimais
        print(f'{chave}: {porcentagem}%')
```
